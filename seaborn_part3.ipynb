```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic=sns.load_dataset('titanic')
titanic.head(5)
```

```python
titanic.columns
```

# Count Plot

```python
titanic.describe()
```

```python
ax=sns.countplot(x='class',data=titanic)
```

```python
ax=sns.countplot(x='class',hue='who',data=titanic)
```

```python
ax=sns.countplot(y='class',hue='who',data=titanic)
```

```python
ax=sns.countplot(x='who',data=titanic,palette='Set3')
```

# Box Plot

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
ax=sns.boxplot(x=tips['total_bill'])
```

```python
ax=sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
ax=sns.boxplot(x='day',y='total_bill'
                ,hue='smoker',data=tips,palette='Set3')
```

```python
ax=sns.boxplot(x='time',y='tip',data=tips
               ,order=['Dinner','Lunch'])
```

```python
iris=sns.load_dataset('iris')
ax=sns.boxplot(data=iris,orient='h',palette='Set2')
```

```python
ax=sns.boxplot(x='day',y='total_bill',data=tips,color='.25')
ax=sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
g=sns.catplot(x='sex',y='total_bill',hue='smoker'
              ,col='time',data=tips,kind='box',height=4,aspect=.7)
```

# Violin Plot

```python
ax=sns.violinplot(x=tips['total_bill'])
```

```python
ax=sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
ax=sns.violinplot(x='day',y='total_bill',hue='smoker'
                  ,data=tips,palette='muted')
```

```python
ax=sns.violinplot(x='day',y='total_bill',hue='smoker'
                  ,data=tips,palette='muted',split=True)
```

```python
ax=sns.violinplot(x='time',y='tip',data=tips,order=['Dinner','Lunch'])
```

```python
ax=sns.violinplot(x='day',y='total_bill',hue='sex',
                  data=tips,palette='Set2',split=True,scale='count')
```

```python
ax=sns.violinplot(x='day',y='total_bill',hue='sex',
                  data=tips,palette='Set2',split=True,scale='count'
                  ,inner='quartile')
```

```python
ax=sns.violinplot(x='day',y='total_bill',hue='sex',
                  data=tips,palette='Set2',split=True,scale='count'
                  ,inner='stick')
```

```python
ax=sns.violinplot(x='day',y='total_bill',hue='sex',
                  data=tips,palette='Set2',split=True,scale='count'
                  ,inner='stick',scale_hue=False,bw=.2)
```

```python
g=sns.catplot(x='sex',y='total_bill',hue='smoker',
              col='time',data=tips,kind='violin',split=True,
              height=4,aspect=.7)
```
