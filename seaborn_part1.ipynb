# KDE Plot (Kernal Density Distribution)

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.kdeplot(data=tips,y='total_bill')
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
sns.kdeplot(data=iris)
```

```python
ax=sns.kdeplot(data=tips,x='total_bill',bw_adjust=0.1)
# bandwidth များရင် smooth ဖြစ်တယ်။
```

```python
ax=sns.kdeplot(data=tips,x='total_bill',bw_adjust=5)
```

```python
ax=sns.kdeplot(data=tips,x='total_bill',bw_adjust=5,cut=0)
# cut zero ထားရင် min/max ကနေစပြီး စဉ်းစားမှာဖြစ်တယ်။
```

```python
sns.kdeplot(data=tips,
            x='total_bill',
            hue='time')
```

```python
sns.kdeplot(data=tips,
            x='total_bill',
            hue='time',
            multiple='stack')
```

```python
tips_agg=(tips
          .groupby('size')
          .agg(total_bill=('total_bill','mean'),
           n=('total_bill','count'))         
)
sns.kdeplot(data=tips_agg,x='total_bill',weights='n')
```

```python
diamonds=sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.kdeplot(data=diamonds,x='price',log_scale=True)
```

```python
ax=sns.kdeplot(data=tips,x='total_bill',
            hue='size',fill=True,alpha=.5,linewidth=0)
```

```python
geyser=sns.load_dataset('geyser')
geyser.head()
```

```python
sns.kdeplot(data=geyser,x='waiting',y='duration')
```

```python
sns.kdeplot(data=geyser,x='waiting',y='duration',hue='kind')
```

```python
sns.kdeplot(data=geyser,x='waiting',y='duration',hue='kind',fill=True)
```

```python
sns.kdeplot(data=geyser,x='waiting',y='duration',
            hue='kind',levels=7,thresh=0.2)
```

```python
ax=sns.kdeplot(data=geyser,x='waiting',y='duration',fill=True,
            levels=100,thresh=0,cmap='mako')
```
